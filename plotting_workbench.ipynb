```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x_qt = 100
y_qt = 100
c = x_qt * y_qt
price_of_x = 2

s = 10000
price_vector = np.array([1, price_of_x])
price_vector = price_vector*s

def utility_curve(x, c):
    y = c / x
    return y

x = np.linspace(0.1, 2000, 1000)
y = utility_curve(x, c)
plt.plot(x, y, label=f"$xy = {c}$")
plt.title("Utility curve and market price")
plt.quiver(0, 0, *price_vector, angles='xy', scale_units='xy', scale=1, color='orange', width=0.003, label=f"$y = px$, $p = {price_of_x}$")
plt.scatter([np.sqrt(c/price_of_x)], [np.sqrt(price_of_x*c)], s=50, color='darkgreen', label="Equilibrium of liquidity pool")
plt.scatter([x_qt], [y_qt], s=50, color='red', label="Initial liquidity pool")

window_size = 300
plt.xlim(0, window_size)
plt.ylim(0, window_size)
plt.legend()
plt.grid()
```
